# **物質的組成**
### 物質的分類
* **純物質**：只由一種物質構成，具有固定的組成與性質，例如水的熔、沸點固定。
  * **元素（單質）**僅包含一種原子的純物質，不能以化學方法再分解成更簡單的物質。
  * 例如：石墨（$C$）、鑽石（$C$）、氧（$O_2$）、硫($S_8$)、鐵($Fe$)$\cdot$等物質。
  1. 目前已知元素有一百多種，可依導電性分爲金屬、非金屬和類金屬三大類。
  > **天然元素**：
<br> **人造元素**：原子序99以後的元素，利用高能粒子互相撞擊而得到，大多不穩定，存在時間相當短，也較難和其他元素反應。
  2. **同素異形體**：由同一種元素所形成的不同物質。

## 原子與分子
* 質量守恆定律
  1. 提出者：西元1789年，法國科學家拉瓦節。so-called近代化學之父。
  2. 内容：**在化學變化發生時，反應前後的總質量相等。**$$HgO\rightarrow Hg + \frac{1}{2}O_2$$$$_{^*本例子是氧化汞的氧化還原反應oxidation-reduction（redox）}$$

* 定比定律（定組成定律）
  1. 提出者：西元1799年，法國科學家普魯斯特
  2. 内容：**某一個化合物，無論其來源或製備方法爲何，其組成元素間都有一定的質量比**。
    * 例：只要化合物為水，則其組成元素氫與氧的質量比必爲1：8。
  3. **定比鼎爐確認了化合物與混合物的差異**：
    1. 化合物中各組成元素具有一定質量比，但混合物的組成則可以改變。
    2. 化合物的性質與其元素不同，但混合物則仍保有其組成成分的性質。

* 倍比定律
  1. 提出者：1804年，道爾吞
  2. 兩種元素可以生成兩種或兩種以上的化合物時，其中一個元素質量固定，另一元素的質量之間會呈簡單整數比。
  3. 當時道爾吞所做的實驗：
    * 已知碳和氧之間有兩種反應，分別是：$$C+O_2\to CO_2$$ $$C+\frac{1}{2}O_2\to CO$$ 用100g碳燃燒$\to$如果燃燒掉133g的氧氣，就會產生$CO$如果燃燒266g氧氣，就會產生$CO_2$。所以：$$1:2=133:266$$
  * 例：

* 氣體化和體積定律與亞佛加厥學說
  1. 給呂薩克-氣體化合體積定律
    * 同溫、同壓下，在氣體反應中，反應物與生成物的反應體積閒呈簡單整數比，但此定律無法以原子説説明。比如1體積的氫+1體積的氯$\to$2體積的氯化氫：$H+Cl\to2H_{\frac{1}{2}}Cl_{\frac{1}{2}}$<br>當然今天我們知道該式應該寫成$H_2+Cl_2\to2HCl$
  2. 亞佛加厥（1811年）
    * 同溫、同壓下，同體積的任何氣體均含有相同數目的分子。假設一公升氯氣$N_2$、氫氣$H_2$、氬氣$Ar$、甲烷$CH_2$，它們都擁有1單位莫爾的分子。

### **原子量**
1. 它是比較質量$\to$沒有單位
  * 原子量 $\neq$ 質量數 = 質子數 + 中子數（質量數必爲正整數）
  * 例：$^{13}C$的質量數為上標13，其原子量為13.003355，兩者并不相等
2. $^{12}_6C=12.0000$
3. 其他原子的原子量是和$^{12}C$比較的結果
  * 例：$^{24}_{12}Mg$質量恰好是$^{12}C$的兩倍。所以$^{24}Mg = 12.0000 \times 2 = 24.0000$

### **平均原子量**
* 元素周期表中所顯示的原子量（比如碳C的原子量是12.01）是碳元素的加權平均。比如$^{12}C$（6個質子，6個中子）自然界中佔的比例為98.9%，$^{13}C$（6個質子，7個中子）佔的比例為1.1%，$^{14}C$（6個質子，8個中子）佔的比例小於0.1%。那麽碳元素的原子量就等於：$$12.0000\times98.9\%+13.003355\times1.1\% \approx12.011$$再比如氫在周期表中的原子量為1.008。但是$^1_1H=1.0078$（占99.99%），也是因爲$^2_1H=2.014$（0.01%）,$^3_1H$（微量）。

### **原子的質量（以amu與克爲單位）**
$$^{12}C=12.000000$$
$$^{13}C=13.003355$$
1. 以莫爾數計：1mol的${12}C$質量為12.0000克，1mol的$^{13}C$質量爲13.003355克。$1mol=6.02\times10^{23}$個碳原子的質量。
2. 個數計：1個$^{12}C$的質量是$\frac{12.0000}{6.02\times10^{23}}$g。這個可以記為12.0000amu。（Atomic Mass Unit）
* amu的定義：$\frac{1}{12}^{12}C$的質量。所以$$1amu=\frac{1}{6.02\times10^{23}}g$$

#### 分子量
* 分子中所含原子的原子量總和
* 例如：已知$^{12}C=12.0000$、$^{12}O=16.0000$可知$CO_2=1\times12+2\times16=44$所以，1個$CO_2$分子的質量為44amu。也可以説1mol的$CO_2$分子的質量為44g。

#### 求分子量的四種方法
1. 分子中所含原子的原子量總和。
  * 例：硫酸$H_2SO_2\times1+1\times32+4\times16=98$其中（H=1，S=32，O=16）
  * 例2：二氧化氮$NO_2=1\times14+2\times16=46$
2. 氣體的分子量（g/mol）=密度（g/L）$\times$莫耳體積（L/mol）= STP（Standard Temperature Pressure。即0$^\circ$C，1atm）下的體積密度$\times$22.4(l/mol)=NTP（Normal TP， 25$^\circ$C, 1atm）下的氣體密度$\times$24.5（l/mol）
3. 亞佛加厥學説的應用。
  * 同溫同壓下，同體積不同氣體莫爾數相同。
    * 例如： 0$^\circ$C，1atm，$O_2$的密度為1.43g/L<br>$O_2$分子量=1.43$\times$22.4=32（g/mol）

### （混合物）平均分子量
* 有混合物，内涵3種物質A（分子量$M_A$，莫爾數的所占比例a%）、物質B（分子量$M_B$，所占比例b%）、物質C（分子量$M_C$，所占比例c%）。<br>其平均分子量為：$M_A\times a\%+M_B\times b\%+M_C\times c\%$（加權平均數）
  * 例：空氣：$\approx80\% N_2 + 20\% O_2$。所以空氣的平均分子量$\approx28\times80\%+32\times20\%=28.8$

### 莫耳與亞佛加厥數
* 1 mole(莫爾)=$6.02\times10^{23}$個（亞佛加厥數）
* 12g的$^{12}C$中所含的原子數（放射衰變、油滴實驗、法拉第定律、晶體繞射）

#### 溶液
* 溶液的組成
  1. 溶液是一種混合物，由兩種以上的物質均匀混合而成。可為固液氣體。
    * 溶液 = 溶劑 + 溶質
  2. 溶劑：
    * 溶液中較多的物質
    * 溶液中有水，無論多寡，水皆爲溶劑。
  3. 溶質：
    * 一般來説，為較少量的物質
    1. 例：少量的鹽加入水中成爲鹽水。則鹽是溶質，水是溶劑，鹽水是溶液。
    2. 例：藥用酒精，95%究竟，5%水。水是溶劑，酒精是溶質。

* 溶液的種類
  1. 依溶劑區分：1.水溶液(aq, aqueous solution)2.非水溶液
  2. 依導電性區分：1.電解質溶液2.非電解質溶液
    * 酸、堿、鹽為電解質
    * 非電解質：有機物（有機酸、氨酚除外）
  3. 依存在的相（三態）分類
    * 溶劑的相決定溶液的相（固體溶劑叫固態溶液）
    * 固態溶於固態：鋅溶於銅$\to$黃銅（合金）
    * 液體溶於固體：汞溶於銀$\to$銀汞齊（牙醫用來補牙）
    * 氣體溶於固體：氫溶於鉑$\to H_2/Pt$
    * 固態溶於液體：碘溶於酒精$\to$碘酒
    * 液體溶於液體：酒精溶於水$\to$酒
    * 氣體溶於液體：$CO_2$溶於水$\to$汽水
    * 固態溶於氣體：碳粒溶於空氣$\to$黑烟
    * 液體溶於氣體：小水滴溶於空氣$\to$雲、霧（白色的都是小水滴、水蒸氣是看不到的）
    * 氣體溶於氣體：氧氣溶於氮氣$\to$空氣

#### 重量百分濃度
* 每100克溶液所含的溶質
  * 100克糖水含有25克糖，濃度？25/100*100%=0.25%

#### 體積百分濃度
* （化學不常用）酒類濃度常用（比如威士忌40度）
* 如：每100毫升溶液中所含的溶質毫升數
  * 300ml啤酒，酒精濃度5%，300*5%=15ml酒精
  $$體積百分濃度=\frac{溶質體積}{溶液體積}，其中，溶液體積\neq 原溶質體積+原溶劑體積$$
  * 20ml酒精+80ml水$\neq$100ml酒

#### 體積莫爾濃度
* 1升溶液中所含的溶質莫爾數（單位M或mol/L）
* $C_M=\frac{溶質莫爾數(mol)}{溶液體積(L)}$，其中，C表示Concentration
* 例，100毫升氯化鈉溶液有0.2mol的氯化鈉$100mlNaCl_{(aq)}$，請問體積莫爾濃度：$$C_M=\frac{0.2}{0.1}=2M$$

### PPM（parts per million） 百萬分濃度
* $10^6$克溶液中所含的溶質克數；（如果在$10^6水中恰有1克DDT，那就是1ppm）；在一般時候，也用$10^3$克溶液中所含溶質的毫克數。

> 以上濃度的共同點$\frac{溶質}{溶液}$

### 溫度如何影響濃度
* 體積受溫度影響
* 重量比較不受溫度影響。所以重量百分濃度和ppm比較不受影響。

  4. 用吸量管將水加至所需的體積。

  1. $C_M=\frac{溶質mol}{溶液體積l}=\frac{x}{0.1}=1，所以x=0.1mol$
  2. 質量/重量=0.1$\times$40=4g
  3. 取4gNaOH溶於適量的水中
  4. 把水加至所需的刻度

#### 飽和溶液vs.不飽和溶液
* 飽和溶液
  * 定義；在定溫下，溶液無法再溶解更多溶質。即溶解的溶質已達最大量，此時溶液為飽和溶液。
  * 特性：為動態平衡->物質溶解的速率==物質結晶的速率。一邊溶解一邊結晶
  * 在飽和溶液中繼續加入溶質，溶液的濃度不會再增加，但假如的溶質形狀會改變

* 不飽和溶液
  * 濃度低於飽和溶液
  * 特性：加入溶質可以持續溶解，溶液濃度持續升高，知道飽和爲止。

* 過飽和溶液
  * 溶液的濃度大於飽和溶液
  * 特性：是不穩定的狀態，攪拌、搖晃，加入少量晶體（可為溶質，也可以不是）。過量溶質會析出，直到濃度與飽和溶液相同。
  * 配製：若一溶質高溫時溶解度大於低溫，則於高溫配製飽和溶液，並趁高溫時過濾雜質和未溶的溶質。
  * 靜置，冷卻，不可搖晃，降溫后即為過飽和溶液。（70$^\circ$C飽和$\to$50$^\circ$C過飽和溶液）
  * 生活中的應用：人造雨。->雲是飽和水蒸氣，創造過飽和->讓水蒸氣降溫->用飛機在雲層上灑下乾冰降溫->再加上AgI碘化銀，析出水珠，下雨。

#### 溶解度
* 溶解度=飽和溶液的濃度
* 定義:固定溫度、壓力下，溶液中所能溶解的溶質最大量，稱爲此溫度下溶質的溶解度。
* 單位：
  * 體積莫爾濃度（M）：1升溶液中所含溶質莫爾數
  * 溶解度分類：>0.1M，稱爲可溶。$10^4$M<微溶< 0.1M。小於$10^4$M，難溶。
  * 100克溶劑中所含的溶質克數。以“溶質克/100克溶劑”表示。
    * 如：25$^\circ$C時100g水可溶40g硝酸鉀表示爲：40克硝酸鉀/100克水

* 影響溶解度的三大因素
  1. 溶質和溶劑的本性
  2. 溫度
  3. 壓力

* 溶質、溶劑本性：同類互溶原則。
1. 極性易溶於極性。水是極性溶劑（通常爲分子不對稱型），易溶解乙酸、乙醇等極性較大的溶質，不易溶解溴（$Br_2$），油脂等非極性溶質。（油水不溶）
2. 非極性容易溶於非極性。如四氯化碳$CCl_4$是非極性溶劑，一溶解溴（$Br_2$），油脂等非極性溶質。水溶性物質通不過細胞膜？因爲有脂雙層

* 溫度因素：
  * 固體：
    * 溶解為吸熱的（較多），溫度上升，溶解度上升
    * 溶解為放熱的（較少），溫度上升，溶解度下降（如硫酸鈣$CaSO_4$、硫酸鈉$Na_2SO_4$）
    * NaCl、KCl等的溶解度不容易受到溫度影響
  * 氣體：溫度上升，溶解度下降

* 壓力因素：
  1. 壓力對固體、液體溶質影響很小。
  2. 氣體溶質：定溫下，壓力越高，溶解度越高。
  3. 亨利定律：對難溶氣體而言，定溫下，一定量溶劑能容的氣體質量正比於其分壓。

#### 拉塞幅的原子核模型
* 葡萄乾布丁原子模型（負電部分均匀部分，剩下的為正電）
